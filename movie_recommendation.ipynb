```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']] # only keep these
movies.head()
```

```python
import ast
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head(2)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head(2)
```

```python
def convert2(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert2)
movies.head(2)
```

```python
def getDirector(text):
    L = []
    for i in ast.literal_eval(text):
      if i['job'] == 'Director':
        L.append(i['name'])
    return L
```

```python
data = {
    'crew': [
        '[{"job": "Director", "name": "John Doe"}, {"job": "Producer", "name": "Jane Smith"}]',
        '[{"job": "Producer", "name": "Jane Doe"}]',
        '[{"job": "Director", "name": "James Bond"}, {"job": "Writer", "name": "Ian Fleming"}]'
    ]
}

m = pd.DataFrame(data)
m['director'] = m['crew'].apply(getDirector)
print(m)
```

```python
movies = movies.drop('crew', axis=1)
```

```python
movies.head(2)
```

```python
movies.to_csv('modified_movies.csv', index=False)
```

```python
movies=pd.read_csv('/content/modified_movies.csv')
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split()) #convert string to list
```

```python
movies.head(2)
```

```python
#removing spaces so that confusion does not occur
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head(2)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast']
```

```python
new_df= movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
new_df[new_df['title'] == 'Avatar'].index[0]
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('John Carter')
```
